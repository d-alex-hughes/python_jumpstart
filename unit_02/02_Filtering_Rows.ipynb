### Find Stops on Divisidero

```python
# Lowercase a string column
df['location'] = df['location'].astype(str) # change the column to a string
df['location'] = df['location'].str.lower()
```

```python
divisadero_stops = df.loc[df['location'].str.contains('divisadero')]
divisadero_stops = len(divisadero_stops)
print(f"There were {divisadero_stops} stops along divisadero st")
```
